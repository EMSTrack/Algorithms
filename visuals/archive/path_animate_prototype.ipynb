```python
# Test the Path class

# Starting Arguments
p1 = Path(Point(0,0), Point(1,1), pieces = 50, display = 5)
print(p1)
```

```python
p3 = Path(Point(0,0), Point(1,1), pieces = 10, display = 5)
```

```python
while True:
    r =  p2.display_points()
    if r: print(r, "\n")
    else: break
```

```python
i = 0
while True:
    
    r =  p3.display_points()
    if r: 
        print(r, "\n")
        plt.figure(i)
        plt.xlim(0, 1)
        plt.ylim(0, 1)
        i += 1
        plt.scatter(
            [p.longitude for p in r],
            [p.latitude for p in r]
        )
    else: break
plt.show()
```
